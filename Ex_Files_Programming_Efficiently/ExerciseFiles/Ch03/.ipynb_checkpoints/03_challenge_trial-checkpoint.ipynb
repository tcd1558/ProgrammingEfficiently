## 03 Challenge

Make an interactive image editor to adjust the brightness, contrast, and saturation of an image

* Matplotlib: https://matplotlib.org

```python
import matplotlib
import matplotlib.pyplot as pp
import matplotlib.animation as anim

import PIL
import PIL.Image

import bs4
```

```python
frame = PIL.Image.open('frames/20120606_000013_512_0171.jpg', 'r')
```

```python
pp.imshow(frame)
```

```python
figure = pp.figure(figsize=(512/72, 512/72), dpi=72)

axes = pp.Axes(figure, [0., 0., 1., 1.])
axes.set_axis_off()
figure.add_axes(axes)

pp.imshow(frame)
```

```python
# Saturation is adjusted as color
def adjust(b,c,s):
    return PIL.ImageEnhance.Color(
                PIL.ImageEnhance.Contrast(
                    PIL.ImageEnhance.Brightnes(frame).enhance(b)
                ).enhance(c)
            ).enhance(s)
```

```python
import ipywidgets
```

```python
# brightness
brightness_control = ipywidgets.FloatSlider(value=1, min=0.5, max=2.0, step=0.05, continuous_update=False)
# contrast
contrast_control = ipywidgets.FloatSlider(value=1, min=0.5, max=2.0, step=0.05, continuous_update=False)
# saturation
saturation_control = ipywidgets.FloatSlider(value=1, min=0.5, max=2.0, step=0.05, continuous_update=False)
```

```python
ipywidgets.interact(pp.imshow(frame), b=brightness_control, c=contrast_control, s=saturation_control)
```

```python
figure = pp.figure(figsize=(512/72, 512/72), dpi=72)

axes = pp.Axes(figure, [0., 0., 1., 1.])
axes.set_axis_off()
figure.add_axes(axes)

pp.imshow(frame)
```
