## 02_07: Olympic medalist challenge

```python
import collections
```

```python
Medal = collections.namedtuple('Medal', ['year','athlete','team','event'])
medals = [Medal(*line.strip().split('\t')) for line in open('goldmedals.txt', 'r')]
```

Challenge:  
Find the four athletes that won the most medals.   
Find the athletes that won medals in at least 4 disciplines

### Find the four athletes that won the most medals

```python
medals[:10]
```

```python
athletes_set = sorted({medal.athlete for medal in medals})
```

```python
athletes_set[:10]
```

Above code finds all athletes by having used a set to filter out duplicate athletes. 
Only athletes, that have won a medal are listed in medals.  
Instead of weeding out the duplicates, leave them in and then count the occurances

```python
athletes_list = sorted([medal.athlete for medal in medals])
```

```python
athletes_list[:10]
```

How do you count the frequency of each entry?  
Use Dictionary Comprehension
https://www.geeksforgeeks.org/python-count-occurrences-element-list/


```python
occurrence = {athlete: athletes_list.count(athlete) for athlete in athletes}
print(occurrence.get('Abebe Bikila'))
```

```python
print(occurrence.get('Abel Kiviat'))
```

```python
occurrence
```

Now sort the dictionary by its value

```python
occurrence.items()
```

```python
sorted(occurrence.items(), key=lambda item: item[1])[-4:]
```

concatenate into a one liner:

```python
sorted({athlete: athletes_list.count(athlete) for athlete in athletes}.items(), key=lambda item: item[1])[-4:]
```

concatenate athletes_list

```python
sorted({athlete: [medal.athlete for medal in medals].count(athlete) for athlete in athletes}.items(), key=lambda item: item[1])[-4:]
```

Is that really more legible?
