## 02_07: Olympic medalist challenge

```python
import collections
```

```python
Medal = collections.namedtuple('Medal', ['year','athlete','team','event'])
medals = [Medal(*line.strip().split('\t')) for line in open('goldmedals.txt', 'r')]
```

Challenge:
Find the four athletes that won the most medals.
Find the athletes that won medals in at least 4 disciplines

### Find the four athletes that won the most medals

Hint:    
    Use collections.Counter   
    Use the method most_common

```python
medals_by_athlete = collections.Counter(medal.athlete for medal in medals)
```

```python
medals_by_athlete
```

```python
medals_by_athlete.most_common(4)
```

### Find the athletes that won medals in at least 4 disciplines

```python
# create an empty dictionary using the type set. 
# The collections.defaultdict creates an entry, if the entry does not exist and assigns it a default value
# The default value below is an empty set
events_by_athlete = collections.defaultdict(set)
```

```python
# fill the empty dictionary with 
for medal in medals:
    events_by_athlete[medal.athlete].add(medal.event)
    # To add to a list you use the method append
    # To append to a dictionary, you use the method add
```

```python
events_by_athlete
```

Some events are listed as gender neutral and gender specific.    
These are the same events.    
Therefore the words 'men' and 'women' can be stripped away.

```python
# Loop over the athletes (keys) and process the value (events)
for athlete,events in events_by_athlete.items():
    print("Key: ", athlete)
    print("Value: ", events)
    for event in events:
        for word in event.split():
            if word not in ['men','women']:
                print(word, end=' ')
        print(',', end=' ')
    print()
```

```python
def normalize(event):
    return ' '.join(word for word in event.split() if word not in ['men', 'women'])
    
```

```python
normalize('discus throw men')
```

```python
events_by_athlete = collections.defaultdict(set)
```

```python
for medal in medals:
    events_by_athlete[medal.athlete].add(normalize(medal.event))
```

```python
events_by_athlete
```

```python
{athlete for athlete, events in events_by_athlete.items() if len(events) >= 4 }
```
